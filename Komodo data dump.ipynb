```python
import pandas as pd
import numpy as np
import requests
```

```python
path = 'https://pinc.activestate.com/licenses/all_komodo/'
response = requests.get(path)
komodo_dump = pd.read_json(response.content, convert_dates=True)
```

```python
# komodo_dump['expiration_date'] = pd.to_datetime(komodo_dump['expiration_date'])
komodo_dump['issued_date'] = pd.to_datetime(komodo_dump['issued_date'])
komodo_dump['updates_until_date'] = pd.to_datetime(komodo_dump['updates_until_date'])
```

```python
criteria = (komodo_dump['updates_until_date'] < np.datetime64('2019-12-31')) & \
           (komodo_dump['updates_until_date'] > np.datetime64('2018-09-01'))
result = komodo_dump[criteria]
result
```
